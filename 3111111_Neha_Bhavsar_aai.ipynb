# AI Research Agents

In this project, we are exploring the latest landscape of **Agentic AI**—specifically, how Multi-Agent systems can be orchestrated to perform deep, autonomous internet research.

Rather than relying on a single AI prompt, this notebook demonstrates a complete, automated assembly line of specialized AI agents working together to investigate the **2026 market adoption rates of Large Language Models (LLMs) in autonomous systems.**

## Agent Architecture
This pipeline utilizes advanced LLMs (like Llama 3.3 and Gemini 3.1) equipped with live web-search capabilities. The workflow is divided into three distinct roles:
1. **The Literature Researcher:** Scours the live internet for up-to-date context, research papers, and hard market statistics.
2. **The Data Analyst:** Reads the raw research to synthesize qualitative market trends and extract critical quantitative data points.
3. **The Report Writer:** Compiles the insights and data into a final, executive-level Markdown document.

# LangChain

````python
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain.agents import create_agent
from langchain_core.tools import tool
from langchain_community.tools import DuckDuckGoSearchRun
from langchain_experimental.tools import PythonREPLTool
from langchain_groq import ChatGroq

# Initialize the model
llm = ChatGoogleGenerativeAI(
    model="gemini-3.1-flash-lite-preview",
    api_key=userdata.get('GOOGLE_API_KEY'),
    temperature=0.2
)
#llm = ChatGroq(model="llama-3.3-70b-versatile", temperature=0)

# Define Tool
search_tool = DuckDuckGoSearchRun(
    name="Literature_Search",
    description="Search the internet for literature, research papers, and up-to-date information."
)

python_analysis_tool = PythonREPLTool(
    name="Data_Analysis",
    description="A Python shell. Use this to execute python commands for data analysis or math."
)

@tool
def write_report(content: str, filename: str = "final_research_report.md") -> str:
    """Write the synthesized research and data analysis into a markdown file."""
    with open(filename, 'w') as f:
        f.write(content)
    return f"Success: Report saved to {filename}"

tools = [search_tool, python_analysis_tool, write_report]

# System Prompt
system_prompt = """You are an elite, autonomous AI Research Assistant and Data Analyst.
Your goal is to complete the user's request by following these steps:

1. Conduct Literature Research: Use the Literature_Search tool to find up-to-date data, context, and statistics.
2. Synthesize & Analyze (Use your brain): Read the findings. Identify key market trends, extract important qualitative insights, and highlight critical numerical statistics.
3. Quantitative Analysis (Optional): ONLY use the Data_Analysis (Python) tool if you need to perform complex mathematical calculations (like averages or percentages) on the numbers you found.
4. Write the Report: Format your qualitative insights, trends, and statistics into a comprehensive, well-structured markdown report.
5. Save: Save the report using the write_report tool.

Always ensure your final report is executive-level, highly detailed, and contextualizes the data points discovered."""

# Agent Creation
agent = create_agent(
    model=llm,
    tools=tools,
    system_prompt=system_prompt
)
````

````python
# 5. Execute the Task
user_query = """
Research the current state of large language models used in autonomous agents as of early 2026.
Find at least 3 key statistics or metrics about their performance or adoption.
Use python to calculate the average of any numerical statistics you find.
Finally, generate a structured markdown report with your findings and save it as 'LLM_Agent_Research.md'.
"""

print("Running research agent... (this may take a minute depending on the search depth)\n")

# The new invocation format requires a list of messages
response = agent.invoke({
    "messages": [{"role": "user", "content": user_query}]
})

print("Process Complete! Final Agent Output:\n")
print(response['messages'][-1].content)
````

````python
from google.colab import userdata
from crewai import Agent, Task, Crew, Process, LLM
from crewai.tools import tool
from langchain_community.tools import DuckDuckGoSearchRun

# llm = LLM(
#     model="groq/llama-3.3-70b-versatile",
#     temperature=0.2
# )
llm = LLM(
    model="gemini/gemini-3.1-flash-lite-preview",
    api_key = userdata.get('GOOGLE_API_KEY'),
    temperature=0.2
)

# Tool Definitions
langchain_search = DuckDuckGoSearchRun()

@tool("Literature_Search")
def search_literature(query: str) -> str:
    """Searches the internet for literature, research papers, and statistics."""
    try:
        return langchain_search.invoke(query)
    except Exception as e:
        return f"Search failed: {e}"

@tool("Calculate_Average")
def calculate_average(numbers: list[float]) -> float:
    """Calculates the average of a list of numbers extracted from data."""
    if not numbers: return 0.0
    return sum(numbers) / len(numbers)

@tool("Save_Report")
def save_report(content: str) -> str:
    """Saves the final markdown report to a file."""
    with open("crew_report.md", 'w') as f:
        f.write(content)
    return "Report successfully saved to crew_report.md."

# Agent Creation
researcher = Agent(
    role="Senior Literature Researcher",
    goal="Find the most up-to-date statistics and context regarding {topic}.",
    backstory="You are a meticulous researcher who scours the web for factual data, ensuring you grab both the numbers and the context surrounding them.",
    tools=[search_literature],
    llm=llm,
    verbose=True
)

analyst = Agent(
    role="Data Analyst",
    goal="Analyze the research findings to identify key trends, important qualitative insights, and extract critical numerical statistics.",
    backstory="You are a seasoned data analyst. You excel at synthesizing raw research into meaningful trends, identifying the most important qualitative context alongside the hard quantitative data points.",
    tools=[calculate_average], # Kept just in case the math is needed
    llm=llm,
    verbose=True
)

writer = Agent(
    role="Technical Report Writer",
    goal="Synthesize research and analysis into a highly structured, professional markdown report and save it.",
    backstory="You format complex data and analytical insights into highly readable, executive-level markdown documents.",
    tools=[save_report],
    llm=llm,
    verbose=True
)

# Task Creation
research_task = Task(
    description="Search the web for current data, context, and at least 3 key statistics about {topic}. Be specific with your search queries.",
    expected_output="A thorough summary of the findings, including the context of the topic, 3 distinct statistics, and their sources.",
    agent=researcher,
)

analysis_task = Task(
    description="Thoroughly review the researcher's findings. Identify the main trends, extract the most important qualitative insights, and highlight the key numerical statistics. If calculating an average of the stats adds value, use your tool to do so.",
    expected_output="A comprehensive analysis detailing key trends, qualitative context, and extracted numerical statistics.",
    agent=analyst,
    context=[research_task]
)

writing_task = Task(
    description="Format the qualitative insights, trends, and statistics from the analyst into a professional markdown report. Use your tool to save the file.",
    expected_output="Confirmation that the report was saved.",
    agent=writer,
    context=[analysis_task]
)
````

````python
# Execute
if __name__ == "__main__":
    crew = Crew(
        agents=[researcher, analyst, writer],
        tasks=[research_task, analysis_task, writing_task],
        process=Process.sequential,
        verbose=True
    )

    topic = "the adoption rate of large language models in autonomous agents"
    result = crew.kickoff(inputs={"topic": topic})
    print("\n--- CREWAI PROCESS COMPLETE ---")
````

````python
import asyncio
from typing import Annotated
from google.colab import userdata
from langchain_community.tools import DuckDuckGoSearchRun

from autogen_agentchat.agents import AssistantAgent
from autogen_agentchat.teams import RoundRobinGroupChat
from autogen_agentchat.conditions import TextMentionTermination, MaxMessageTermination
from autogen_agentchat.ui import Console
from autogen_ext.models.openai import OpenAIChatCompletionClient

# Used Lanngchain Tool - There is Error with Serper API
langchain_search = DuckDuckGoSearchRun()

def search_literature(query: Annotated[str, "The search query to look up online"]) -> str:
    """Searches the internet for literature, research papers, and statistics."""
    try:
        return langchain_search.invoke(query)
    except Exception as e:
        return f"Search failed: {e}"

def calculate_average(numbers: Annotated[list[float], "List of numbers to average"]) -> str:
    """Calculates the mathematical average of a list of numbers."""
    if not numbers: return "No numbers provided."
    avg = sum(numbers) / len(numbers)
    return f"The calculated average is: {avg}"

def save_report(content: Annotated[str, "The complete markdown formatted report"]) -> str:
    """Saves the final report to a markdown file."""
    with open("autogen_report.md", 'w') as f:
        f.write(content)
    return "Report successfully saved to autogen_report.md"


async def main():
    # LLM config
    # model_client = OpenAIChatCompletionClient(
    #     model="llama-3.3-70b-versatile",
    #     api_key=userdata.get("GROQ_API_KEY"),
    #     base_url="https://api.groq.com/openai/v1",
    #     model_info={
    #         "vision": False,
    #         "function_calling": True,
    #         "json_output": True,
    #         "family": "llama",
    #         "structured_output": False,
    #     },
    # )
    # Load Gemini Key
    model_client = OpenAIChatCompletionClient(
        model="gemini-3.1-flash-lite-preview",
        api_key=userdata.get('GOOGLE_API_KEY'),
        base_url="https://generativelanguage.googleapis.com/v1beta/openai/", # <-- Google's OpenAI endpoint
        model_info={
            "vision": False,
            "function_calling": True,
            "json_output": True,
            "family": "unknown"
        },
    )

    # Agents Creation (The Chat Room Participants)
    researcher_agent = AssistantAgent(
        name="Literature_Researcher",
        system_message="""You are Step 1. Your job is to use the 'search_literature' tool to find current data, context, and at least 3 key statistics about the user's topic.
        Once you have the data, summarize your findings and pass the conversation to the Data_Analyst. DO NOT calculate anything.""",
        model_client=model_client,
        tools=[search_literature],
    )

    analyst_agent = AssistantAgent(
        name="Data_Analyst",
        system_message="""You are Step 2. Read the research provided by the Literature_Researcher.

        Your job is twofold:
        1. LITERATURE SYNTHESIS: Extract the most important qualitative context, historical background, and major industry trends from the text. Explain the 'why' and 'how'.
        2. DATA EXTRACTION: Highlight the key numerical statistics.

        If calculating an average of those statistics adds value, use the 'calculate_average' tool.
        Pass your comprehensive analysis (both the literature context and the numbers) cleanly to the Report_Writer.""",
        model_client=model_client,
        tools=[calculate_average],
    )

    writer_agent = AssistantAgent(
        name="Report_Writer",
        system_message="""You are Step 3. Read the analysis from the Data_Analyst.
        Format it into a highly structured, professional Markdown report.
        Call the 'save_report' tool with the exact Markdown content.
        Once the tool returns a success message, you MUST output the exact word TERMINATE to end the workflow.""",
        model_client=model_client,
        tools=[save_report],
    )

    # Team & Orchestration (Round Robin: Researcher -> Analyst -> Writer)
    termination = TextMentionTermination("TERMINATE") | MaxMessageTermination(8)

    team = RoundRobinGroupChat(
        [researcher_agent, analyst_agent, writer_agent],
        termination_condition=termination,
    )

    # Execute
    user_query = "Research the current state of large language models used in autonomous agents as of early 2026. Find 3 statistics about their adoption."

    print("\n--- AUTOGEN CONVERSATION STREAMING ---")

    await Console(team.run_stream(task=user_query))

    print("\n--- AUTOGEN PROCESS COMPLETE ---")


if __name__ == "__main__":
    # Leave alone this error it is cause by how colab runtime works ,it will not couse you any problem
    await main()
````
